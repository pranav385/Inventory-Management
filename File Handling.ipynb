```python
# Function to get a valid phone number from the user
def get_phone_number():
    while True:
        u_phone_no=input("Enter Phone Number - ")
        if u_phone_no.isdigit() and int(u_phone_no)>0:
            u_phone_no = u_phone_no
            return u_phone_no
        else:
            print("Please enter a valid phone number.")

# Function to create a dictionary from the product list with Product ID as key and quantity as value
def product_dict(products):
    pset={}
    for i in products:
        fi=i.split(',')
        pset[fi[0]]=int(fi[3])
    return pset

# Function to get a valid Product ID from the user
def user_Pid(pset):
    while True:
        u_id=input("Enter Product ID - ")
        if u_id in pset and int(pset[u_id])>0:
            u_id=u_id
            break
        else:
            print('Please write Product ID which is in the inventory or Product is out of stock')
    return u_id

# Function to get a valid quantity from the user
def user_PQuantity():
    while True:
        u_Quantity=input('Enter Quantity you want - ')
        if u_Quantity.isdigit() and int(u_Quantity)>0:
            u_Quantity=u_Quantity
            break
        else:
            print('Please Enter a valid quantity')
    return u_Quantity

# Reading products from 'Inventory.txt'
fd=open('Inventory.txt','r')
products=fd.read().split('\n')
products
fd.close()

# Displaying products to the user
for p in products:
    print(p)
    
print('-'*60)
print('Please Enter your Detail for Exciting offers!')

# Taking user details
u_name=input('Enter Your Name - ')
print('Hi ',u_name,'!')
print('If you give us your Phone Number, You will get exciting offer. We will respect your Privacy.')
print()

# Asking the user if they want to provide a phone number
ph=input('Press y to give Phone Number or any key to continue without Phone Number - ')

if ph in 'yY':
    u_phone_no=get_phone_number()
            
else:
    u_phone_no='Not Given'
    print('Phone Number not given')


print('-'*60)

print('Please choose products you want and in how many quantity')

# Creating a dictionary from the product list
pset = product_dict(products)

# Getting Product ID and Quantity from the user
u_id = user_Pid(pset)
u_Quantity = user_PQuantity()

print('-'*60)

updated_product_list=[]

# Processing the order
for pr in products:
    fpr=pr.split(',')
    if u_id == fpr[0] and int(fpr[3])>0:
        
        if int(u_Quantity)<=int(fpr[3]):
            # Displaying the details of the selected product and calculating billing amount
            print('Product ID :'+'           '+fpr[0])
            print('Product Name :'+'         '+fpr[1])
            print('Quantity :'+'             '+u_Quantity)
            print('Price  :'+'               '+fpr[2])
            print('Billing Amount :'+'       '+str(int(u_Quantity)*int(fpr[2])))
            pay=int(u_Quantity)*int(fpr[2])
            print('-'*60)
            discount=pay*0.1
            # Applying a discount if a phone number is provided
            if u_phone_no != 'Not Given':
                print('As you provided your Phone Number, so 10% discount applied')
                print('Your total Payable discount is :- ',str(pay-discount))
            else:
                print('If you give your phone number you have to pay',str(pay-discount),'Otherwise',str(pay))
                
                # Asking the user if they want to provide a phone number after seeing the total payable amount
                ph1 = input('If you want to give Phone Number press Y or press any key to continue without Phone Number - ')
                
                if ph1 in 'yY':
                    u_phone_no=get_phone_number()
                    print('Thanks for providing Phone Number. 10% discount applied')
                    print('Your total Payable discount is :- ',str(pay-discount))
                else:
                    print('Your total Payable discount is :- ',str(pay))
            
            # Updating the quantity in the product list
            fpr[3]=str(int(fpr[3])-int(u_Quantity))
            
            # Writing the transaction details to 'Sales.txt'
            fd= open('Sales.txt','a')
            if u_phone_no != 'Not Given':
                detail=u_name +','+u_phone_no +','+fpr[1]+','+u_Quantity+','+ str(pay-discount)+str(pay)+'\n'
            else:
                detail=u_name +','+u_phone_no +','+fpr[1]+','+u_Quantity+','+ str(pay)+str(pay)+'\n'
            fd.write(detail)
            fd.close()
            
        else:
            # If the requested quantity is more than the available quantity
            print('We do not have that much Quantity. We have maximum',fpr[3],'Quantity.','Do you want to proceed and buy',fpr[3],'Quantities')
            chr = input('Press Y to continue or N to quit')
            
            if chr in 'yY':
                # Displaying the details for the available quantity and calculating billing amount
                print('Product ID :'+'           '+fpr[0])
                print('Product Name :'+'         '+fpr[1])
                print('Quantity :'+'             '+fpr[3])
                print('Price  :'+'               '+fpr[2])
                print('Billing Amount :'+'       '+str(int(fpr[3])*int(fpr[2])))
                print('-'*60)
                pay=int(fpr[3])*int(fpr[2])
                discount=pay*0.1
                
                # Applying a discount if a phone number is provided
                if u_phone_no != 'Not Given':
                    print('As you provided your Phone Number, so 10% discount applied')
                    print('Your total Payable discount is :- ',str(pay-discount))
                else:
                    print('If you give your phone number you have to pay',str(pay-discount),'Otherwise',str(pay))

                    # Asking the user if they want to provide a phone number after seeing the total payable amount
                    ph1 = input('If you want to give Phone Number press Y or press any key to continue without Phone Number - ')

                    if ph1 in 'yY':
                        u_phone_no=get_phone_number()
                        print('Thanks for providing Phone Number. 10% discount applied')
                        print('Your total Payable discount is :- ',str(pay-discount))
                    else:
                        print('Your total Payable discount is :- ',str(pay))

                # Updating the quantity in the product list
                fpr[3]=str(0)
                
                # Writing the transaction details to 'Sales.txt'
                fd= open('Sales.txt','a')
                if u_phone_no != 'Not Given':
                    detail=u_name +','+u_phone_no +','+fpr[1]+','+fpr[3]+','+ str(pay-discount)+str(pay)+'\n'
                else:
                    detail=u_name +','+u_phone_no +','+fpr[1]+','+fpr[3]+','+ str(pay)+str(pay)+'\n'
                fd.write(detail)
                fd.close()
            else:
                print('Sorry for the inconvinence, Buy other product from our store. Thanks')
    updated_product_list.append(fpr)
    
print('Thanks, Plz Visit Again')

# Updating the 'Inventory.txt' file with the updated product list
lst=[]
for i in updated_product_list:
    updated_Inventory=((",").join(i)+"\n")
    lst.append(updated_Inventory)
lst[-1]=lst[-1][:-1] ## Removing extra line which is inserted while adding last product detail
fd=open("Inventory.txt",'w')
for i in lst:
    fd.write(i)
fd.close()

print('-'*60)
print('Inventory Updated')

# Displaying the updated inventory
fd=open('Inventory.txt','r')
products=fd.read()
fd.close()
print(products)
```
